```python
# 8. Detección de valores inusuales (ejemplo para una columna numérica)
print("\n=== VALORES INUSUALES (EJEMPLO) ===")
sample_col = 'yearsatcompany'  # Puedes cambiar por cualquier columna numérica
q1 = df_datos_abc2[sample_col].quantile(0.25)
q3 = df_datos_abc2[sample_col].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
outliers = df_datos_abc2[(df_datos_abc2[sample_col] < lower_bound) | (df_datos_abc2[sample_col] > upper_bound)]
print(f"Valores inusuales en {sample_col}:")
print(outliers[sample_col].unique())
```

```python
# 9. Análisis de formato para columnas de texto
text_cols = df_datos_abc2.select_dtypes(include=['object']).columns
print("\n=== ANÁLISIS DE FORMATO (TEXTOS) ===")
for col in text_cols:
    sample_values = df_datos_abc2[col].dropna().sample(min(5, len(df_datos_abc2[col])))
    print(f"\nMuestra de valores en {col}:")
    print(sample_values.values)
```
