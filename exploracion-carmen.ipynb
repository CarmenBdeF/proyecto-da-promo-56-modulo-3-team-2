```python
#copia de la variable df_carmen para no borrar nada definitivamente
df_limpio_carmen = df_carmen.copy()
print("Iniciando el proceso de limpieza de datos...")
```

```python
#1. Columna 'numberchildren'
# Esta columna no contiene información. Valorar si le recomendamos a la empresa tener o no esa info disponible.
print("Limpiando 'numberchildren'...")
if 'numberchildren' in df_limpio_carmen.columns:
    df_limpio_carmen.drop('numberchildren', axis=1, inplace=True)
    print("'numberchildren' ha sido eliminada.")
else:
    print("'numberchildren' no se encontró en el DataFrame.")
```

```python
#Comprobamos que se ha eliminado correctamente
df_limpio_carmen.columns
```

```python
# 2. Columna 'yearsincurrentrole'
# Es de tipo 'object' y los números usan una coma como separador decimal. Tiene una gran cantidad de valores nulos.
print("\nLimpiando 'yearsincurrentrole'...")
df_limpio_carmen['yearsincurrentrole'] = df_limpio_carmen['yearsincurrentrole'].str.replace(',', '.', regex=False).astype(float)
print("'yearsincurrentrole' convertida a tipo numérico (float).")
# NOTA: Esta columna tiene una cantidad muy alta de valores nulos.
# Deberás decidir una estrategia para manejarlos, como rellenarlos (con la media, mediana, etc.) o incluso eliminar la columna si no es útil para tu análisis.
```

### Ya tenemos yearsatcurrenrole cambiado a float. Falta ver qué hacer con los nulos (amplia mayoría. No le aplicaría la media o mediana puesto que son demasiados nulos y no veo tendencia. La respuesta está en usar yearssincelastpromotion y yearsatcompany.

```python
# 3. Columnas de moneda: 'sameasmonthlyincome' y 'salary'
# Ambas son de tipo 'object', contienen el símbolo '$' y usan comas para los decimales.
print("\nLimpiando 'sameasmonthlyincome' y 'salary'...")

def limpiar_moneda(series_moneda):
    if pd.api.types.is_object_dtype(series_moneda):
        # Aseguramos que todos los valores son strings antes de usar .str
        return series_moneda.astype(str).str.replace('$', '', regex=False).str.replace(',', '.', regex=False).astype(float)
    return series_moneda

df_limpio_carmen['sameasmonthlyincome'] = limpiar_moneda(df_limpio_carmen['sameasmonthlyincome'])
df_limpio_carmen['salary'] = limpiar_moneda(df_limpio_carmen['salary'])
print("Columnas de moneda ('sameasmonthlyincome', 'salary') convertidas a tipo float.")
```

```python
df_limpio_carmen.describe().T #ahora también tenemos sameasmonthlyincome y salary como float
```

```python
df_limpio_carmen.info()
```

```python
# 4. Columna 'roledepartament'
# Tiene espacios en blanco extra y mayúsculas/minúsculas inconsistentes.
print("\nLimpiando 'roledepartament'...")
df_limpio_carmen['roledepartament'] = df_limpio_carmen['roledepartament'].str.strip().str.title()
print("'roledepartament' normalizada (sin espacios extra y en formato Título).")
```

### Podemos ver que roledepartament es en realidad una combinación de las columnas jobrole y department, por lo que esta columna es redundante. Sugiero limpiar y completar la columna department (basándonos en jobrole) y borramos roledepartment. La estrategia de nulos aquí es entonces con imputación (dict)

```python
# 5. Columna 'remotework'
# Contiene valores booleanos representados de distintas formas (Yes, 1, True, 0, False).
# Vamos a estandarizarlos a un único tipo booleano (True/False).
print("\nLimpiando 'remotework'...")
mapa_booleano = {
    'Yes': True, '1': True, 1: True, 'True': True, True: True,
    'False': False, '0': False, 0: False, 'No': False, False: False
}
df_limpio_carmen['remotework'] = df_limpio_carmen['remotework'].map(mapa_booleano).astype(bool)
print("'remotework' estandarizada a tipo booleano.")
```
