Deberías investigar qué variables pueden influir en la satisfacción laboral y en la rotación del personal. Estas son algunas columnas clave:

Demográficos:
Age — edad del empleado
Gender — género
MaritalStatus — estado civil
NUMBERCHILDREN — número de hijos (si está disponible)

Profesionales:
JobRole — puesto específico del empleado
Department — departamento en el que trabaja
JobLevel — nivel jerárquico del puesto
Education, EducationField — nivel educativo y área de formación
YearsAtCompany, YearsInCurrentRole, YearsSinceLastPromotion, YEARSWITHCURRMANAGER — antigüedad en la empresa, en el puesto, desde la última promoción, y con el mismo gerente
OverTime — si realiza horas extras
BusinessTravel — frecuencia de viajes por trabajo

Económicos y de beneficios:
MonthlyIncome — ingreso mensual
HourlyRate, DailyRate, MonthlyRate — tarifas de remuneración
PercentSalaryHike — incremento porcentual del salario
StockOptionLevel — nivel de opciones sobre acciones otorgadas

Rendimiento y desarrollo profesional:
PerformanceRating — evaluación de desempeño
TrainingTimesLastYear — cantidad de formaciones recibidas el último año
NumCompaniesWorked — número de empresas anteriores
TotalWorkingYears — años totales de experiencia laboral
